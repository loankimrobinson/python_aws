```python
import boto3 #pip3 install --proxy=http://proxy-server.bms.com:8080/ boto3
import pandas as pd
```

```python
import os
os.environ["AWS_DEFAULT_REGION"] = 'us-east-2'
os.environ["AWS_ACCESS_KEY_ID"] = 'youraccesskey'
os.environ["AWS_SECRET_ACCESS_KEY"] = 'yoursecretaccesskey'
```

```python
s3 = boto3.resource(
    service_name= 's3',
    region_name = 'us-east-2',
    aws_access_key_id = 'youraccesskey',
    aws_secret_access_key = 'yoursecretaccesskey'
)
```

```python

# get samples.csv.gz from the bucket

bucket = s3.Bucket("yourbucket")
obj = s3.Object("yourbucket", 'samples.csv.gz')
df = pd.read_csv(obj.get()['Body'], compression='gzip', sep = '\t')
df.head(5)
```

```python
# list all of object in the bucket with prefix
bucket = s3.Bucket("yourbucket")
prefix = 'yourprefixlink' 

dt_objects = []
for index, objects in enumerate(bucket.objects.filter(Prefix=prefix), start=1):   # default is zero
    dt_objects.append((index, objects.key))
    
dt_objects = pd.DataFrame(dt_objects,columns=('Index', 'Location'))    
dt_objects
```
